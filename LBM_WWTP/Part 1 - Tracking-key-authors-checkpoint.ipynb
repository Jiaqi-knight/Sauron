## Towards modelling  biotechnological wastewater treatment processes through lattice Boltzmann method - Part 1

This is the first part of a 3 notebook series where the main goal was to review the scientific bibliography related to the mathematical modelling of biotechnological wastewater treatment processes through the lattice Boltzmann method. Through a literature screening, we found 72 papers reporting this issue and, in order to track the main authors and keywords of these studies, it was created this notebook.

### Checking python version

```python
from platform import python_version
print(python_version())
```

```python
import os
import wordcloud
from os import path
from wordcloud import WordCloud
import pandas as pd
import matplotlib.pyplot as plt
```

###  Tracking key-Authors

```python
text = open(path.join('key-authors.txt')).read()
type(text)
```

### Getting number of the words in .txt file

```python
words = text.split(' ')
print('Number of words in text file :', len(words))
```

### Generating a word cloud object

```python
wordcloud = WordCloud(collocations=False, 
                      background_color="white", 
                      max_words=200, width=1500, height=1000,
                      random_state=1).generate(text)
plt.figure(figsize = (15,15))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

According to the figure above, we observe that Picioreanu and Loosdrecht are the main authors in this field of research, followed by Heijnen and Yang.

###  Tracking key-Words

```python
word = open(path.join('words_cloud.txt')).read()
type(word)
```

### Getting number of the words in .txt file

```python
words = word.split(' ')
print('Number of words in text file :', len(words))
```

```python
### Generating a word cloud object
```

```python
wordcloud = WordCloud(collocations=False, 
                      background_color="white", 
                      max_words=400, width=1500, height=1000,
                      random_state=1).generate(word)
plt.figure(figsize = (15,15))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

According to the figure above, we observe that **'model' 'biofilm' 'flow'** were the main keywords, followed by **'method' 'lattice Boltzmann'**. 
This leads us to conclude that the majority of the studies have been applied LBM to model biofilm related-processes, mainly fluid flow behaviour.
